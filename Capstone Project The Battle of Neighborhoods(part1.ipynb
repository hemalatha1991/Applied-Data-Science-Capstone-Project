# Capstone Project - The Battle of Neighborhoods (Week 1)

#### The Battle of Neighborhoods: Introduction/Business Problem

#### Introduction

This project is part of Applied Data Science capstone project. In this project I am going to provide recommendation for our client to open Dutch restaurant in Toronto area. Dutch cuisine is famous for their fries, waffels, stompot, pancakes etc. So, my client want to expand their service in Canada. With that in mind finding the suitable location for my client to open restaurant is the important for my client so I am creating this project to help my client to find the best location. 

#### Problem statement

The main objective of this Applied Data Science capstone project is to provide recommendation to my client to open Ducth restaurant in Toronto area,Canada. For this I am going to use Clustering approach to solve this problem. This project aims to address this business question: If my client want to open Dutch restaurant in Toronto, In Which neighbourhood/area/location should they open it?

#### Target Audience

My client wants to find the location to open typical Dutch restaurant.

#### Data

For this project I am going to use these below mentioned dataset:
    1. List of neighbourhood in Toronto,Canada from wikipedia.
    2. Latitude and longitude of the neighbourhood from Geocoder package.
    3. Location/venue data related to Dutch restaurant from Foursqaure.

I am going to do the analysis, results and conclusion in the week 2 project.
